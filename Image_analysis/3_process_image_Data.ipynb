```python
## let's change the csv output from fiji

#Full inflorescence pictures: 346 Four bract pictures: 349 One bract pictures: 349
#image_num = 346
#df = pd.read_csv('pictures/all_tif/1.full_Whole_Color_Measurements.csv')

#image_num = 349
#df = pd.read_csv('pictures/all_tif/2.four_Whole_Color_Measurements.csv')

image_num = 349
df = pd.read_csv('pictures/all_tif/3.one_Whole_Color_Measurements.csv')

df
```

```python
## there's issues with the HSB values. Not all of them have 3. Let's just remove anything that has _HSB

no_HSB_df = df[df['Label'].str.endswith(('RGB', 'LAB'))]
no_HSB_df

#no_HSB_df.to_excel("no_HSB.xlsx")
```

```python
pic = "one"
Colorspace = [pic + "_R", pic + "_G", pic + "_B", pic + "_L*", pic + "_A*", pic + "_B*"]
#Colorspace = ["RGB_R", "RGB_G", "RGB_B", "LAB_L", "LAB_A", "LAB_B"]
no_HSB_df['Type'] = Colorspace*image_num
no_HSB_df
```

```python
## in order for this to work, make sure that the ones labeled a and b are shortened somehow. either that or figure
# out how to slice by anything between "F2-06-" and "_"

# Group by the first two characters of 'Name' and calculate the mean of 'Age'
mean_df = no_HSB_df.groupby([no_HSB_df['Label'].str[:9],no_HSB_df['Type']])['Mean1'].mean().reset_index()

# Rename the columns for clarity
mean_df.columns = ['Name_Start', 'Type', 'Mean_Value']

mean_df
```

```python
mean_df_full = mean_df.pivot(index = 'Name_Start', columns = 'Type', values = 'Mean_Value')
mean_df_full = mean_df_full.reset_index()
mean_df_full
```

```python
mean_df_four = mean_df.pivot(index = 'Name_Start', columns = 'Type', values = 'Mean_Value')
mean_df_four = mean_df_four.reset_index()
mean_df_four
```

```python
mean_df_one = mean_df.pivot(index = 'Name_Start', columns = 'Type', values = 'Mean_Value')
mean_df_one = mean_df_one.reset_index()
mean_df_one
```

```python
result = pd.merge(mean_df_full, mean_df_four, on='Name_Start', how='inner')


mean_df_all = pd.merge(result, mean_df_one, on='Name_Start', how='inner')
mean_df_all
```

```python
mean_df_all.at[50, 'Name_Start'] = "F2-06-220A"
mean_df_all.iloc[50]
```

```python
mean_df_all['Name_Start'] = mean_df_all['Name_Start'].str.replace('_', '')
mean_df_all.to_excel("pictures/all_tif/image_analysis.xlsx", index=False)
mean_df_all
```

```python
df = pd.read_excel('colorimeter print sheet.xlsx', sheet_name = "format to combine")
color_df = df.drop(columns = ['Bract #'])
color_df
```

```python
# Group by Tree Id and Bract #
mean_color_df = color_df.groupby('Tree ID').mean().reset_index()
mean_color_df.columns = ['Name_Start', "color_L", "color_a", "color_b"]
mean_color_df
```
